# The Drunkard's Walk

## 1. Background
- The drunkard is at 0.
- The drunkard moves either right or left only.
- If the probability that the drunkard moves right is p, which location will be the drunkard at?

## 2. Rules
- The drunkard moves n times.

```python
import matplotlib.pyplot as plt

import collections
import random
```

```python
%matplotlib inline
```

```python
# M : trials
# N : total moves
# p : probability that drunkard moves right
M = 10000
N = 20
p = 0.6
```

```python
drunkard_positions = list()
for i in range(M):
    position = 0
    for j in range(N):
        if random.random() < p: position += 1
        else: position -= 1
    drunkard_positions.append(position)
```

```python
title_fontsize, label_fontsize, tick_fontsize = 25, 23, 13

for trial in [100, 1000, 10000]:
    plt.figure(figsize=(20, 10))
    x, y = list(), list()
    for a, b in sorted(collections.Counter(drunkard_positions[:trial]).items(), key=lambda x: x[0]):
        x.append(a)
        y.append(b)
    plt.bar(x, y)
    plt.title('Trials : ' + str(trial), fontsize=title_fontsize)
    plt.xlabel('Position', fontsize=label_fontsize)
    plt.ylabel('Counts', fontsize=label_fontsize)
    plt.xticks(range(-N, N+1, 2), fontsize=tick_fontsize)
    plt.yticks(fontsize=tick_fontsize)
plt.show()
```
