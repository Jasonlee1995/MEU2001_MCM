# Calculating Pi

## 1. Background
- Use random to get Pi.
- Use Riemann Sum to get Pi.

## 2. Rules
- Use r=1 circle to get Pi.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
```

```python
# Use random
trials = 1000
pi_random = list()

for trial in range(1, trials):
    matrix = np.random.uniform(-1, 1, (trial, 2))
    x, y = matrix[:, 0], matrix[:, 1]
    in_circle = np.sum((np.power(x, 2)+np.power(y, 2)) <= 1)
    pi_random.append(4*in_circle/trial)
```

```python
# Use Riemann Sum
sections = 1000
pi_riemann = list()

for section in range(1, sections):
    x = np.arange(0, 1, 1/section)
    y = np.sqrt(1-np.power(x, 2))
    pi_riemann.append(4*np.sum(y)/section)
```

```python
print('Pi original : {:.6f}'.format(np.pi))
print('Pi using random : {:.6f}'.format(np.mean(pi_random)))
print('Pi using Riemann Sum : {:.6f}'.format(pi_riemann[-1]))
```

```python
label_fontsize = 25

plt.figure(figsize=(20, 10))
plt.plot(pi_random)
plt.plot([np.pi]*trials, color='red')
plt.xlabel('Trial', fontsize=label_fontsize)
plt.ylabel('Pi_random', fontsize=label_fontsize)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
plt.plot(pi_riemann)
plt.plot([np.pi]*sections, color='red')
plt.xlabel('Trial', fontsize=label_fontsize)
plt.ylabel('Pi_riemann', fontsize=label_fontsize)
plt.show()
```
