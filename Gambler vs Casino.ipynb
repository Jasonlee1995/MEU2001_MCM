# Gambler vs Casino

## 1. Background
Gambler has 100\\$ and casino has 2000\\$ each.  
If the probability that the gambler will win over the casino in the game is p, then what is the actual probability that the gambler will win all the money the casino has?

## 2. Rules
- If the gambler wins the game, he gets \\$1. If he loses the game, on the contrary, he loses \\$1. (The same applies to the casino.)
- The game will not end until either the gambler or the casino loses all the money.

```python
import matplotlib
import matplotlib.pyplot as plt

import random
```

```python
%matplotlib inline
```

```python
# P : probability that the gambler will win over the casino
# N : trials
P = 0.501
N = 10
```

```python
gambler_wins, plays = [], []
gambler_budget_traces = []

for i in range(N):
    gambler_budget, casino_budget = 100, 2000
    gambler_budget_trace = []
    trial = 0
    while (gambler_budget != 0) and (casino_budget != 0):
        trial += 1
        gambler_budget_trace.append(gambler_budget)
        if random.random() < P: gambler_budget, casino_budget = gambler_budget+1, casino_budget-1
        else: gambler_budget, casino_budget = gambler_budget-1, casino_budget+1
    if gambler_budget == 0: gambler_wins.append(0)
    else: gambler_wins.append(1)
    gambler_budget_traces.append(gambler_budget_trace)
    plays.append(trial)
```

```python
print('Probability that the gambler will win over the casino :', sum(gambler_wins)/len(gambler_wins))
```

```python
label_fontsize, tick_fontsize = 23, 13

plt.figure(figsize=(20, 10))
graph = plt.bar(range(N), plays)

# if gambler win, bar color is blue or not bar color is red
for i in range(N):
    if gambler_wins[i] == 1: graph[i].set_color('b')
    else: graph[i].set_color('r')

plt.xlabel('Trial', fontsize=label_fontsize)
plt.ylabel('Number of plays', fontsize=label_fontsize)
plt.xticks(range(N), fontsize=tick_fontsize)
plt.yticks(fontsize=tick_fontsize)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
for i in range(N):
    plt.plot(gambler_budget_traces[i])
    plt.xlabel('Number of plays', fontsize=label_fontsize)
    plt.ylabel("Gambler's budget ($)", fontsize=label_fontsize)
    plt.xticks(fontsize=tick_fontsize)
    plt.yticks(fontsize=tick_fontsize)
plt.show()
```
