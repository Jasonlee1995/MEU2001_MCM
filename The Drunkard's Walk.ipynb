# The Drunkard's Walk

## 1. Background
- The drunkard is at 0.
- The drunkard moves either right or left only.
- If the probability that the drunkard moves right is p, which location will be the drunkard at?

## 2. Rules
- The drunkard moves n times.

```python
import matplotlib
import matplotlib.pyplot as plt

import collections
import random
```

```python
%matplotlib inline
```

```python
# M1, M2, M3 : trials
# N : total moves
# p : probability that drunkard moves right
M1, M2, M3 = 100, 1000, 10000
N = 20
p = 0.6
```

```python
drunkard_position1, drunkard_position2, drunkard_position3 = list(), list(), list()
for i in range(M3):
    position = 0
    for j in range(N):
        if random.random() < p: position += 1
        else: position -= 1
    if i < M1: drunkard_position1.append(position)
    if i < M2: drunkard_position2.append(position)
    drunkard_position3.append(position)
```

```python
title_fontsize, label_fontsize, tick_fontsize = 25, 23, 13

plt.figure(figsize=(20, 10))
y, x, patches = plt.hist(drunkard_position1, range=(-N, N+1), bins=2*N+1, align='left')

for i in range(0, len(y)):
    if y[i] != 0:
        plt.text(x=x[i]-0.3, y=y[i]+0.2,
                 s=str(int(y[i])),
                 fontsize=tick_fontsize, color='black')

plt.title('M1', fontsize=title_fontsize)
plt.xlabel('Position', fontsize=label_fontsize)
plt.ylabel('Trial', fontsize=label_fontsize)
plt.xticks(range(-N, N+1), fontsize=tick_fontsize)
plt.yticks(fontsize=tick_fontsize)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
y, x, patches = plt.hist(drunkard_position2, range=(-N, N+1), bins=2*N+1, align='left')

for i in range(0, len(y)):
    if y[i] != 0:
        plt.text(x=x[i]-0.3, y=y[i]+0.2,
                 s=str(int(y[i])),
                 fontsize=tick_fontsize, color='black')

plt.title('M2', fontsize=title_fontsize)
plt.xlabel('Position', fontsize=label_fontsize)
plt.ylabel('Trial', fontsize=label_fontsize)
plt.xticks(range(-N, N+1), fontsize=tick_fontsize)
plt.yticks(fontsize=tick_fontsize)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
y, x, patches = plt.hist(drunkard_position3, range=(-N, N+1), bins=2*N+1, align='left')

for i in range(0, len(y)):
    if y[i] != 0:
        plt.text(x=x[i]-0.3, y=y[i]+0.2,
                 s=str(int(y[i])),
                 fontsize=tick_fontsize, color='black')

plt.title('M3', fontsize=title_fontsize)
plt.xlabel('Position', fontsize=label_fontsize)
plt.ylabel('Trial', fontsize=label_fontsize)
plt.xticks(range(-N, N+1), fontsize=tick_fontsize)
plt.yticks(fontsize=tick_fontsize)
plt.show()
```
